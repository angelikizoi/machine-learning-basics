## Decision Tree on Categorical Data
This example demonstrates how to apply a **Decision Tree Classifier** to a dataset containing categorical features, where we predict whether a customer buys a computer based on their age, income, student status, and credit rating.

```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
# Define the dataset
data = {
    "age": [ '<=30', '<=30', '31..40', '>40', '>40', '>40', '31..40', '<=30', '<=30', '>40', '<=30', '31..40', '31..40', '>40'],
   
    "income":['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium',                                               'medium', 'high', 'medium'],
     
     "student": [ 'no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no' ],
   
     "credit_rating": [ 'fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair',     
                           'excellent', 'excellent', 'fair', 'excellent'],
       
      "buys_computer": [ 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

# Create DataFrame
df = pd.DataFrame(data)
```

### Data Encoding
Since the data is categorical, we use map function to convert the categories into numerical values.

```python
d = { '<=30' : 0 , '31..40' : 1, '>40' : 2}
df['age'] = df['age'].map(d)
 
d = { 'low' : 0, 'medium' : 1, 'high' : 2}
df['income'] = df['income'].map(d)
 
d = {'no': 0, 'yes': 1}
df['student'] = df['student'].map(d)
 
d = {'fair': 0, 'excellent': 1}
df['credit_rating'] = df['credit_rating'].map(d)
 
d = {'yes': 1, 'no': 0}
df['buys_computer'] = df['buys_computer'].map(d)
 
```

### Model Training & Visualization

```python
# Define features (X) and target (y)
features = ['age', 'income', 'student', 'credit_rating']
X = df[features]
y = df['buys_computer']

# Initialize and train the Decision Tree Classifier
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X,y)


# Visualize the decision tree
tree.plot_tree(dtree, feature_names=features)
plt.show()
```
