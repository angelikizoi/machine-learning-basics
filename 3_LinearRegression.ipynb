### Simple Linear Regression with scipy.stats.linregress

This example demonstrates how to perform simple linear regression using scipy.stats.linregress to model the relationship between advertising spending and sales revenue.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Advertising spending (x) and sales revenue (y) data
advertising_spending = np.array([5, 15, 25, 35, 45, 55])
sales_revenue = np.array([5, 20, 14, 32, 22, 38])

# Perform linear regression and get the slope, intercept, and R-squared value
slope, intercept, r_value, p_value, std_err = stats.linregress(advertising_spending, sales_revenue)
print(f"R-Squared: {r_value**2}")

# Define a function to predict sales based on advertising spending
def predict(ads_expense):
    return slope*ads_expense + intercept

# Calculate predicted sales and plot the data with the regression line
predicted_sales = predict(advertising_spending)
plt.scatter(advertising_spending, sales_revenue)
plt.plot(advertising_spending, predicted_sales, c='r')
plt.xlabel('Advertising Spending ($1000)')
plt.ylabel('Sales Revenue ($1000)')
plt.title('Linear Regression (scipy.stats)')
plt.show()
```

### Linear Regression with scikit-learn
The following code demonstrates how to use scikit-learn to perform linear regression on the Diabetes dataset. It includes loading the dataset, splitting the data into training and test sets, fitting a linear regression model and evaluating the model using Mean Squared Error (MSE).

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
# Create the DataFrame needed by loading the Diabetes dataset from sklearn
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['disease_progression'] = diabetes.target
df.head()
```

```python
features = diabetes.feature_names # Independent variables
target = 'disease_progression' # Dependent variable

# Split the dataset into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=0)
```

```python
# Initialize the Linear Regression model
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, y_train)
```

```python
# Predict the target values for the test data
y_pred = model.predict(X_test)

# Calculate the Mean Squared Error (MSE) for the model's predictions
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"Root Mean Squared Error: {rmse:.2f}")
```
