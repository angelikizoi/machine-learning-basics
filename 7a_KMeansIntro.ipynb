# KMeans Clustering - An Introduction
This notebook demonstrates how to perform KMeans clustering and how to use the Elbow method to find the optimal number of clusters for a given dataset.

## Example 1

## Step 1: Import necessary libraries
We start by importing the necessary libraries: `KMeans` from the `sklearn` library for clustering, and `matplotlib` for visualizing the results.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

## Step 2: Initialize the dataset

```python
x = [6, 7, 12, 6, 5, 13, 16 , 8, 12, 14]
y = [20, 18, 23, 16, 15, 24, 25, 21, 20, 20]

data = list(zip(x, y))
```

## Step 3: Use the Elbow Method to find the optimal number of clusters
The Elbow method helps us determine the optimal number of clusters by plotting the sum of squared distances (inertia) against the number of clusters.
We calculate the inertia for a range of cluster numbers (from 1 to 10) and plot it to identify the 'elbow point', where adding more clusters results in diminishing returns.

```python
# Initialize a list to store the inertia values for each number of clusters
inertias = []  # List to hold inertia values for each k
 
# Loop over a range of cluster numbers, from 1 to 10
for i in range(1, 11):  # We will test cluster sizes from 1 to 10
    kmeans = KMeans(n_clusters=i)  # Initialize KMeans with i clusters
    kmeans.fit(data)  # Fit the KMeans model on the dataset
    inertias.append(kmeans.inertia_)  # Append the inertia (sum of squared distances) to the list
 
# Plot the results using matplotlib
plt.plot(range(1, 11), inertias, marker='o')  # Plot inertia vs number of clusters
plt.title('Elbow Method to Determine Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')  # Inertia is the sum of squared distances
plt.show()  # Display the plot
```

In the plot generated above, the 'elbow point' refers to the number of clusters where the inertia starts decreasing at a slower rate. This indicates the optimal number of clusters for the dataset.

## Step 4: Perform KMeans clustering with 2 clusters
Based on the Elbow method, we have determined that the optimal number of clusters for this dataset is 2. Now, we will use KMeans with `n_clusters=2` to cluster the data points.

```python
# Perform KMeans clustering with 2 clusters
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
```

## Step 5: Visualize the clustered data
After fitting the model, we can now visualize the clusters. The data points will be colored according to the cluster labels assigned by the KMeans algorithm.

```python

plt.scatter(x,y,c=kmeans.labels_)
plt.show()
```

Here is the scatter plot showing the result of KMeans clustering with 2 clusters. The points are colored based on the cluster to which they were assigned.

The plot visually distinguishes the two groups found by the KMeans algorithm.

## Example 2

### Step 1: Initialize the dataset and calculate inertias

- **Dataset**: X contains six points in two-dimensional space.

```python
X = np.array([
    [1, 2],
    [1.5, 1.8],
    [5, 8],
    [8, 8],
    [1, 0.6],
    [9, 11]
])
```

- **Inertia Calculation**: For each number of clusters from 1 to 6, the KMeans algorithm is run, and the inertia (sum of squared distances of samples to their closest cluster center) is calculated and stored in inertias.
- **Elbow Plot**: The inertias are plotted against the number of clusters to help decide the optimal number of clusters using the Elbow method.

```python
inertias = []
for i in range(6):
    kmeans = KMeans(n_clusters=i+1)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,7), inertias, marker='*')
plt.show()   
```

### Step 2: Perform KMeans clustering with 2 clusters

- The KMeans algorithm is applied to the dataset with 2 clusters.
- 'centers' contains the coordinates of the two cluster centers found by KMeans.

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
centers = kmeans.cluster_centers_
```

### Step 3: Plot the clustered points and cluster centers

- **Scatter plot**: The data points are plotted and colored according to the cluster they belong to.
- **Cluster centers**: The cluster centers are marked with large red stars (*).

```python
plt.scatter(X[:,0],X[:,1], s=100, c=kmeans.labels_, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], s=300, c='red', marker='*')
plt.show()
```
